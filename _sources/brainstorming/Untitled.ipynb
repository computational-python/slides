```python
# %load plotting.py
from matplotlib import pyplot as plt

from jax import vmap, grad
from jax.numpy import tanh, linspace

tanh_derivative = grad(tanh)

t = linspace(-2, 2)

plt.plot(t, tanh(t))
plt.plot(t, vmap(tanh_derivative)(t))
```
